```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.linspace(-2,2,10000)
```

```python
def g(x):
    return (x**4-(4*x**2)+5);
def dg(x):
    return (4*x**3-8*x);
```

```python
# superimposing gradient decent values on plots
plt.figure(figsize=(55,20))
#plot 1
plt.subplot(1,2,1) #row column and index this is used to define plot 1
plt.title("Cost Function")
plt.plot(data,g(data))
#plot 2
plt.subplot(1,2,2)
plt.title("Slope of cost Function")
plt.grid()
plt.plot(data,dg(data),color = 'red')
```

```python
#gradient decent
```

```python
new_x = 2
previous_x = 0
step_multiplier = 0.001
precision = 0.0001
x_list = [new_x]
slope = [dg(new_x)]
for i in range(500):
    previous_x = new_x
    gradient = dg(previous_x)
    new_x = previous_x - (step_multiplier*gradient)
    step_size = abs(new_x - previous_x)
    x_list.append(new_x)
    slope.append(dg(new_x))
    if step_size < precision:
        print("loop ran {} times".format(i))
        break
print("Local Minima Occurs at: ",new_x)
print("Slope at this Point: ",dg(new_x))
print("F(x) or cost value at this point: ",g(new_x))
```
