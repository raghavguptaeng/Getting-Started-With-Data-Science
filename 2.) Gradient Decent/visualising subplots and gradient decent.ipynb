```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# y = x^2 + x + 1 any arbitrary Function
```

```python
def f(x):
    return x**2+x+1;
```

```python
x_1 = np.linspace(start=-6,stop=6,num=1000)
```

```python
plt.xlim([-6,6])
plt.ylim(0,30)
plt.plot(x_1,f(x_1))
```

```python
#Now lets create sub plots
```

```python
def df(x): #done by differentiation
    return 2*x+1;
```

```python
# plot function and derivative side by side
plt.figure(figsize=(25,10))
#plot 1
plt.subplot(1,2,1) #row column and index this is used to define plot 1
plt.title("Cost Function")
plt.xlim([-6,6])
# plt.xlabel('X')
# plt.ylabel('X')
plt.ylim(0,30)
plt.plot(x_1,f(x_1))
#plot 2
plt.subplot(1,2,2)
plt.title("Slope of cost Function")
plt.grid()
plt.plot(x_1,df(x_1),color = 'red',linewidth = 4)
plt.show()
```

```python
#lets start with gradient Decent
new_x = 6
previous_x = 0
step_multiplier = 0.1
precision = 0.001
x_list = [new_x]
slope_list = [df(new_x)]
for i in range(300):
    previous_x = new_x
    gradient = df(previous_x)
    new_x = previous_x - step_multiplier*gradient
    step_size = abs(new_x - previous_x)
    x_list.append(new_x)
    slope_list.append(df(new_x))
    if step_size<precision:
        print('loop ran {0} times'.format(i))
        break
print("Local Minima Occurs at: ",new_x)
print("Slope at this Point: ",df(new_x))
print("F(x) or cost value at this point: ",f(new_x))
#here we can see that values of local minima are appox and their accuracy depends on number of time loop is running. however it never reaches -0.5 which is its true value
#we can also see that our slope approaches more towards 0 and our f(x) is accurate as it should be 
#refer to notes attached
```

```python
# superimposing gradient decent values on plots
plt.figure(figsize=(25,10))
#plot 1
plt.subplot(1,2,1) #row column and index this is used to define plot 1
plt.title("Cost Function")
plt.xlim([-6,6])
# plt.xlabel('X')
# plt.ylabel('X')
plt.ylim(0,30)
values = np.array(x_list)
plt.scatter(x_list,f(values),color='black',alpha=0.3)
plt.plot(x_1,f(x_1))
#plot 2
plt.subplot(1,2,2)
plt.title("Slope of cost Function")
plt.grid()
plt.plot(x_1,df(x_1),color = 'red')
plt.scatter(x_list,slope_list,color='black',alpha=0.3)
plt.show()
```
